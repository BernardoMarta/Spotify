```python
import pandas as pd
import plotly.express as px
```

```python
df_2013_2023=pd.read_json('Spotify Extended Streaming History\\Streaming_History_Audio_2013-2023_0.json')
print(df_2013_2023.__len__())
df_2013_2023.head()
```

```python
df_2023_2025=pd.read_json('Spotify Extended Streaming History\\Streaming_History_Audio_2023-2025_1.json')

print(df_2023_2025.__len__())
df_2023_2025.head()
```

```python
df_2025_2=pd.read_json('Spotify Extended Streaming History\\Streaming_History_Audio_2025_2.json')
print(df_2025_2.__len__())

df_2025_2.head()
```

```python
df_2022_2025_video=pd.read_json('Spotify Extended Streaming History\\Streaming_History_Video_2022-2025.json')

df_2022_2025_video.head()
```

```python
df_music = pd.merge(df_2013_2023, df_2023_2025, 'outer')

df_music = pd.merge(df_music, df_2025_2, 'outer')

print((df_2013_2023.__len__()+ df_2023_2025.__len__()+df_2025_2.__len__() ))
df_music.__len__()
```

```python
df_music.info()
```

```python
df_top_music = df_music[['ts','ms_played','master_metadata_track_name','master_metadata_album_album_name','master_metadata_album_artist_name','episode_name','episode_show_name']]
df_top_music = df_top_music.rename(
    columns={
        'ts':'timestamp',
        'ms_played':'milisseconds',
        'master_metadata_track_name': 'track',
        'master_metadata_album_album_name': 'album',
        'master_metadata_album_artist_name': 'artist',
        'episode_name': 'episode',
        'episode_show_name': 'podcast'
    }
)
```

```python
df_top_music[['date','time']] = df_top_music['timestamp'].str.split('T',expand=True)
df_top_music['time']=df_top_music[['time']].replace('Z$','',regex=True)
df_top_music['month']=df_top_music['date'].str.split('-').str[1]
df_top_music['hour']=df_top_music['time'].str.split(':.*').str[0]
df_top_music.head()
```

```python
df_top_music['minutes']=df_top_music['milisseconds']/(1000*60)
df_top_music['type']=df_top_music['episode'].isna().map({True:'Music', False:'Podcast'})
df_top_music=df_top_music[['timestamp','minutes','date','month','time','hour','artist','album','track', 'podcast', 'episode','type']]
df_top_music
```

```python
counts = df_top_music.groupby(['hour','type'])['minutes'].sum().reset_index(name='count')

counts['count'] = counts['count'].round(0)

hour_fig = px.bar(
    counts,
    x = 'hour',
    y = 'count',
    labels={'x': 'Hours', 'y':'Number of musics played' },
    title = 'Music played by hour',
    color='type',
    barmode='group'
)

hour_fig.update_layout(
    title='Musics and Podcasts listened per Hour',
    title_x=0.5,
    xaxis_title = 'Hour',
    yaxis_title = 'Number of Minutes',
    legend = dict(
        title = 'Select the information'
    )
)

hour_fig.show()

hour_fig.write_html('hour_fig.html')
```

```python
counts = df_top_music.groupby(['month','type'])['minutes'].sum().reset_index(name='count')

counts['count'] = counts['count'].round(0)

month_fig = px.bar(
    counts,
    x = 'month',
    y = 'count',
    labels={'x': 'Hours', 'y':'Number of musics played'},
    title = 'Music played by hour',
    color='type',
    barmode='group'
)

month_fig.update_layout(
    title='Musics and Podcasts listened per Month',
    title_x=0.5,
    xaxis_title = 'Month',
    yaxis_title = 'Number of Minutes',
    legend = dict(
        title = 'Select the information'
    )
)

month_fig.show()

month_fig.write_html('month_fig.html')
```

```python
import plotly.express as px

# Aggregate minutes by month and type
counts = df_top_music.groupby(['month','type'])['minutes'].sum().reset_index(name='count')
counts['count'] = counts['count'].round(0)

# Create the bar chart with all types visible
month_fig = px.bar(
    counts,
    x='month',
    y='count',
    color='type',
    barmode='group',
    labels={'month':'Month','count':'Minutes listened','type':'Content Type'},
    title='Musics and Podcasts listened per Month'
)

# Create dropdown buttons
buttons = [
    dict(label='All',
         method='update',
         args=[{'visible':[True, True]},
               {'yaxis': {'title':'Minutes listened'}}]),
    dict(label='Music only',
         method='update',
         args=[{'visible':[True, False]},
               {'yaxis': {'title':'Minutes listened'}}]),
    dict(label='Podcast only',
         method='update',
         args=[{'visible':[False, True]},
               {'yaxis': {'title':'Minutes listened'}}])
]

month_fig.update_layout(
    updatemenus=[dict(active=0, buttons=buttons, x=1.15, y=0.9)],
    title_x=0.5
)

month_fig.show()
month_fig.write_html('month_fig.html')
```
